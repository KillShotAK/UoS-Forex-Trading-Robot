```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization, MultiHeadAttention, Lambda, Add
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import KFold
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Ensure TensorFlow uses the GPU, if available
device_name = tf.test.gpu_device_name()
if device_name != '/device:GPU:0':
    print('GPU device not found')
else:
    print('Found GPU at: {}'.format(device_name))

# Set random seeds for reproducibility
def reset_random_seeds(seed=42):
    os.environ['PYTHONHASHSEED'] = str(seed)
    tf.random.set_seed(seed)
    np.random.seed(seed)
    random.seed(seed)

reset_random_seeds()

def preprocess_datetime_columns(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['DayofWeek'] = df['Date'].dt.dayofweek
    df = df.drop(columns=['Date'])
    return df

def preprocess_text_columns(df, text_columns, max_words=5000, max_len=100):
    tokenizer = Tokenizer(num_words=max_words)
    for col in text_columns:
        tokenizer.fit_on_texts(df[col])
        sequences = tokenizer.texts_to_sequences(df[col])
        padded = pad_sequences(sequences, maxlen=max_len)
        df = df.drop(columns=[col])
        padded_df = pd.DataFrame(padded, columns=[f"{col}_word_{i}" for i in range(padded.shape[1])])
        df = pd.concat([df.reset_index(drop=True), padded_df.reset_index(drop=True)], axis=1)
    return df, tokenizer

def build_model(input_shape, dropout_rate=0.4, learning_rate=0.00005, num_heads=8):
    input_data = Input(shape=(input_shape,))

    # Self-Attention
    reshaped_input = Lambda(lambda x: tf.expand_dims(x, axis=1))(input_data)
    attention_output = MultiHeadAttention(num_heads=num_heads, key_dim=min(input_shape, 64))(reshaped_input, reshaped_input)
    attention_output = Lambda(lambda x: tf.squeeze(x, axis=1))(attention_output)

    # Dense layers with BatchNormalization and Dropout
    x = Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(attention_output)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate)(x)
    x = Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate)(x)
    x = Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate)(x)

    # Output Layer
    output = Dense(3, activation='softmax')(x)

    model = Model(inputs=input_data, outputs=output)

    # Compile model with an adjustable learning rate
    model.compile(optimizer=tf.keras.optimizers.AdamW(learning_rate=learning_rate),
                  loss='sparse_categorical_crossentropy',
                  metrics=['sparse_categorical_accuracy'])

    return model

def plot_confusion_matrix(y_pred, y_true, classes):
    cm = confusion_matrix(y_true, np.argmax(y_pred, axis=1))
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.title('Confusion Matrix')
    plt.show()

def train_model(X_train, y_train, X_val, y_val, X_test, y_test, batch_size=32, epochs=200, learning_rate=0.00005, dropout_rate=0.4, num_heads=8):
    model = build_model(X_train.shape[1], dropout_rate, learning_rate, num_heads)

    # Early stopping and learning rate reduction on plateau
    early_stopping = EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=20, min_lr=1e-7)

    history = model.fit(X_train, y_train,
                        validation_data=(X_val, y_val),
                        epochs=epochs,
                        batch_size=batch_size,
                        callbacks=[early_stopping, reduce_lr])

    # Evaluate the model
    loss, accuracy = model.evaluate(X_test, y_test)
    print(f"Test Accuracy: {accuracy}")

    # Generate Confusion Matrix
    y_pred = model.predict(X_test)
    plot_confusion_matrix(y_pred, y_test, classes=['Down', 'Hold', 'Up'])

    # Plot Accuracy and Loss
    plt.plot(history.history['sparse_categorical_accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_sparse_categorical_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    return model

def k_fold_cross_validation(X, y, k=5, **kwargs):
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    histories = []
    accuracies = []

    for fold, (train_index, val_index) in enumerate(kf.split(X), 1):
        # Ensure X and y are numpy arrays if they are not
        X = np.array(X)
        y = np.array(y)

        # Apply SMOTE to balance the dataset
        smote = SMOTE(random_state=42)
        X_resampled, y_resampled = smote.fit_resample(X[train_index], y[train_index])

        # Scale the features
        scaler = StandardScaler()
        X_resampled = scaler.fit_transform(X_resampled)
        X_val_scaled = scaler.transform(X[val_index])

        # Train model on this fold
        model = train_model(X_resampled, y_resampled, X_val_scaled, y[val_index], X_val_scaled, y[val_index], **kwargs)

        # Evaluate model on the validation set
        _, accuracy = model.evaluate(X_val_scaled, y[val_index])
        accuracies.append(accuracy)
        histories.append(model.history.history)

        print(f"Fold {fold} Accuracy: {accuracy}")

    # Print the average accuracy over all folds
    print(f"Average Accuracy over {k}-folds: {np.mean(accuracies)}")
    return accuracies, histories

# Load your data
train_data = pd.read_pickle("/content/X_train.pkl")
test_data = pd.read_pickle("/content/X_test.pkl")
val_data = pd.read_pickle("/content/X_val.pkl")

train_labels = pd.read_pickle("/content/y_train.pkl")
test_labels = pd.read_pickle("/content/y_test.pkl")
val_labels = pd.read_pickle("/content/y_val.pkl")

# Preprocess datetime columns
train_data = preprocess_datetime_columns(train_data)
test_data = preprocess_datetime_columns(test_data)
val_data = preprocess_datetime_columns(val_data)

# Identify and preprocess text columns
text_columns = ['title', 'articleBody']  # Replace with actual column names containing text
train_data, _ = preprocess_text_columns(train_data, text_columns)
test_data, _ = preprocess_text_columns(test_data, text_columns)
val_data, _ = preprocess_text_columns(val_data, text_columns)

# Hyperparameter tuning
dropout_rate = 0.4
learning_rate = 0.00005
batch_size = 32
num_heads = 8
epochs = 200

# Perform k-fold cross-validation with SMOTE and improved model architecture
accuracies, histories = k_fold_cross_validation(train_data.values, train_labels, k=5,
                        batch_size=batch_size, epochs=epochs,
                        learning_rate=learning_rate, dropout_rate=dropout_rate, num_heads=num_heads)

# If the third fold performs the best, fine-tune the model based on the third fold
best_fold = 3
train_model(train_data.values, train_labels, val_data.values, val_labels, test_data.values, test_labels,
            batch_size=batch_size, epochs=epochs,
            learning_rate=learning_rate, dropout_rate=dropout_rate, num_heads=num_heads)
```
